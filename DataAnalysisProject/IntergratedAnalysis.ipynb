```python
#Visualizations
# Create a function to plot a given matrix of the list of players for each seasons.
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] =8,6

def myplot (data = Salary,playerlist = Players):
    cdict = {"KobeBryant":'Black',"JoeJohnson":'Blue',"LeBronJames":'Green',"CarmeloAnthony":'Red',"DwightHoward":'Yellow',"ChrisBosh":'Magenta',"ChrisPaul":'Grey',"KevinDurant":'Cyan',"DerrickRose":'Purple',"DwayneWade":'Indigo'}
    for name in playerlist:
        plt.plot(data[Pdict[name]],c=cdict[name], ls='-',marker = '.', ms=10, label = Players[Pdict[name]])
    plt.legend(loc='upper left',bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons,rotation = 'vertical')
    plt.show()

myplot(Games, ["CarmeloAnthony","ChrisPaul","DerrickRose"] )

# Create a function to plot Free throw attempts per game for the given list of players. Use star marker and dotted line.
def Freethrowattemptspergameplot (playerlist = Players):
    cdict = {"KobeBryant":'Black',"JoeJohnson":'Blue',"LeBronJames":'Green',"CarmeloAnthony":'Red',"DwightHoward":'Yellow',"ChrisBosh":'Magenta',"ChrisPaul":'Grey',"KevinDurant":'Cyan',"DerrickRose":'Purple',"DwayneWade":'Indigo'}
    for name in playerlist:
        plt.plot(FreeThrowAttempts[Pdict[name]]/Games[Pdict[name]],c=cdict[name], ls=':',marker = '*', ms=8, label = Players[Pdict[name]])
    plt.legend(loc='upper left',bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons,rotation = 'vertical')
    plt.show()

Freethrowattemptspergameplot()

# Create a function to plot accuracy of free throws for the given list of players. Use square marker and dashed line.
def Freethrowaccuracyplot (playerlist = Players):
    cdict = {"KobeBryant":'Black',"JoeJohnson":'Blue',"LeBronJames":'Green',"CarmeloAnthony":'Red',"DwightHoward":'Yellow',"ChrisBosh":'Magenta',"ChrisPaul":'Grey',"KevinDurant":'Cyan',"DerrickRose":'Purple',"DwayneWade":'Indigo'}
    for name in playerlist:
        plt.plot(FreeThrows[Pdict[name]]/FreeThrowAttempts[Pdict[name]],c=cdict[name], ls='--',marker = 's', ms=6, label = Players[Pdict[name]])
    plt.legend(loc='upper left',bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons,rotation = 'vertical')
    plt.show()
    
Freethrowaccuracyplot()

# Create a function to plot Player playing style. For this, calculate points per field goals excluding free throws. Each free throw worth 1 point
def Playerplayingstyle (playerlist = Players):
    cdict = {"KobeBryant":'Black',"JoeJohnson":'Blue',"LeBronJames":'Green',"CarmeloAnthony":'Red',"DwightHoward":'Yellow',"ChrisBosh":'Magenta',"ChrisPaul":'Grey',"KevinDurant":'Cyan',"DerrickRose":'Purple',"DwayneWade":'Indigo'}
    for name in playerlist:
        plt.plot((Points[Pdict[name]] - FreeThrows[Pdict[name]])/FieldGoals[Pdict[name]],c=cdict[name], ls='-.',marker = 'h', ms=7, label = Players[Pdict[name]])
    plt.legend(loc='upper left',bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons,rotation = 'vertical')
    plt.show()

Playerplayingstyle()



```
